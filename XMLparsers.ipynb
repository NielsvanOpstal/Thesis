```python
from xml.etree.ElementTree import iterparse
import pandas as pd

## TODO ##########################################

def parsePand(XMLfile, outputFile):
    """
    input: file name of XML file from panden database from BAG
    returns: pandas dataframe that contains data from XML file
    """
    with open(outputFile, "a", encoding="utf-8") as output:
    
        d = {"bouwjaar": "NaN", "pandstatus" : "NaN", "begindatumTijdvakGeldigheid" : "NaN", "einddatumTijdvakGeldigheid" : "NaN", 
            "inOnderzoek" : "NaN", "documentdatum" : "NaN", "documentnummer" : "NaN",
            "identificatie" : "NaN", "aanduidingRecordInactief" : "NaN", "officieel" : "NaN", "posList" : "NaN"}

        for i, (tag, elem) in enumerate(iterparse(XMLfile)):

            if elem.tag == "{http://www.kadaster.nl/schemas/imbag/lvc/v20090901}bouwjaar":

                output.write(d["bouwjaar"] + ";" + d["pandstatus"] + ";" + d["begindatumTijdvakGeldigheid"] + ";" + 
                         d["einddatumTijdvakGeldigheid"] + ";" + d["inOnderzoek"] + ";" + d["documentdatum"] + ";" + d["documentnummer"] + ";" +
                         d["identificatie"] + ";" + d["aanduidingRecordInactief"] + ";" + d["officieel"] + ";" + d["posList"] + "\n")
                
                
                d = {"bouwjaar": "NaN", "pandstatus" : "NaN", "begindatumTijdvakGeldigheid" : "NaN", "einddatumTijdvakGeldigheid" : "NaN", 
                        "inOnderzoek" : "NaN", "documentdatum" : "NaN", "documentnummer" : "NaN",
                        "identificatie" : "NaN", "aanduidingRecordInactief" : "NaN", "officieel" : "NaN", "posList" : "NaN"}


                d["bouwjaar"] = elem.text

            elif elem.tag == "{http://www.kadaster.nl/schemas/imbag/lvc/v20090901}pandstatus":
                d["pandstatus"] = elem.text

            elif elem.tag == "{http://www.kadaster.nl/schemas/imbag/imbag-types/v20090901}begindatumTijdvakGeldigheid":
                d["begindatumTijdvakGeldigheid"] = elem.text

            elif elem.tag == "{http://www.kadaster.nl/schemas/imbag/imbag-types/v20090901}einddatumTijdvakGeldigheid":
                d["einddatumTijdvakGeldigheid"] = elem.text


            elif elem.tag == "{http://www.kadaster.nl/schemas/imbag/lvc/v20090901}inOnderzoek":
                d["inOnderzoek"] = elem.text

            elif elem.tag == "{http://www.kadaster.nl/schemas/imbag/imbag-types/v20090901}documentdatum":
                d["documentdatum"] = elem.text

            elif elem.tag == "{http://www.kadaster.nl/schemas/imbag/imbag-types/v20090901}documentnummer":
                d["documentnummer"] = elem.text

            elif elem.tag == "{http://www.kadaster.nl/schemas/imbag/lvc/v20090901}identificatie":
                d["identificatie"] = elem.text

            elif elem.tag == "{http://www.kadaster.nl/schemas/imbag/lvc/v20090901}aanduidingRecordInactief":
                d["aanduidingRecordInactief"] = elem.text

            elif elem.tag == "{http://www.kadaster.nl/schemas/imbag/lvc/v20090901}officieel":
                d["officieel"] = elem.text

            elif elem.tag == "{http://www.opengis.net/gml}posList":
                d["posList"] = elem.text

```

```python
import time
start_time = time.time()

import os
# assign directory
directory = '../Data/BAG/PAND'
# iterate over files in
# that directory
tot = len(os.listdir(directory))
c = 0
outputName = "PAND.csv"

with open(outputName, "w", encoding="utf-8") as output:
    
    output.write("bouwjaar;pandstatus;begindatumTijdvakGeldigheid;einddatumTijdvakGeldigheid;inOnderzoek;documentdatum;"
                 "documentnummer;identificatie;aanduidingRecordInactief;officieel;posList" + "\n")
    
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    c += 1
    if c % 20 == 0:
        print("{} % done".format(c / tot * 100))
    if os.path.isfile(f):
        parsePand(f, outputName)
print("elapsed time:{} minutes ".format((time.time() - start_time) / 60))
```

```python
from xml.etree.ElementTree import iterparse
import pandas as pd
"""
Pand ID always follows the tag "{http://www.kadaster.nl/schemas/imbag/lvc/v20090901}bron". Multiple instances of the tag for
ID are the same but can refer to for example an adress or pand. Since I'm interested in the building for now I'll only use the
ID referring to the building.

More data can be found in the XML files. Only taking the necesarry.

seperating data using ;
"""
def parseVBO(XMLfile, outputName):
    with open(outputName, "a", encoding = "utf-8") as output:
        
        d = {"gebruiksdoelVerblijfsobject" : "NaN", "oppervlakteVerblijfsobject": "NaN", "verblijfsobjectStatus" : "NaN", "begindatumTijdvakGeldigheid" : "NaN", "einddatumTijdvakGeldigheid" : "NaN", 
            "identificatie" : "NaN", "pos" : "NaN"}
        idcheck = False
        
        for i, (tag, elem) in enumerate(iterparse(XMLfile)): # iterate over lines in XML

            if elem.tag == "{http://www.kadaster.nl/schemas/imbag/lvc/v20090901}gebruiksdoelVerblijfsobject":

                output.write(d["gebruiksdoelVerblijfsobject"] + ";" + d["oppervlakteVerblijfsobject"] + ";" + d["verblijfsobjectStatus"] + ";" + 
                         d["begindatumTijdvakGeldigheid"] + ";" + d["einddatumTijdvakGeldigheid"] + ";" + d["identificatie"] + ";" + d["pos"] + "\n")
                
                d = {"gebruiksdoelVerblijfsobject" : "NaN", "oppervlakteVerblijfsobject": "NaN", "verblijfsobjectStatus" : "NaN", "begindatumTijdvakGeldigheid" : "NaN", "einddatumTijdvakGeldigheid" : "NaN", 
                "identificatie" : "NaN", "pos" : "NaN"}

                d["gebruiksdoelVerblijfsobject"] = elem.text

            elif elem.tag == "{http://www.kadaster.nl/schemas/imbag/lvc/v20090901}oppervlakteVerblijfsobject":
                d["oppervlakteVerblijfsobject"] = elem.text

            elif elem.tag == "{http://www.kadaster.nl/schemas/imbag/lvc/v20090901}verblijfsobjectStatus":  
                d["verblijfsobjectStatus"] = elem.text

            elif elem.tag == "{http://www.kadaster.nl/schemas/imbag/imbag-types/v20090901}begindatumTijdvakGeldigheid":
                d["begindatumTijdvakGeldigheid"] = elem.text

            elif elem.tag == "{http://www.kadaster.nl/schemas/imbag/imbag-types/v20090901}einddatumTijdvakGeldigheid":
                d["einddatumTijdvakGeldigheid"] = elem.text

            elif elem.tag == "{http://www.kadaster.nl/schemas/imbag/lvc/v20090901}bron":
                idcheck = True

            elif elem.tag == "{http://www.kadaster.nl/schemas/imbag/lvc/v20090901}identificatie" and idcheck:
                d["identificatie"] = elem.text
                idcheck = False

            elif elem.tag == "{http://www.opengis.net/gml}pos":
                d["pos"] = elem.text

            
            
```

```python
start_time = time.time()
import os
# assign directory
directory = '../Data/BAG/VBO'
# iterate over files in
# that directory
tot = len(os.listdir(directory))
c = 0
outputName = "VBO.csv"

with open(outputName, "w", encoding = "utf-8") as output:
    output.write("gebruiksdoelVerblijfsobject;oppervlakteVerblijfsobject;verblijfsobjectStatus;begindatumTijdvakGeldigheid;"
                "einddatumTijdvakGeldigheid;identificatie;pos" + "\n")
    
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    c += 1
    if c % 20 == 0:
        print("{} % done".format(c / tot * 100))
    if os.path.isfile(f):
        parseVBO(f, outputName)
        
print("elapsed time:{} minutes ".format((time.time() - start_time) / 60))
```
