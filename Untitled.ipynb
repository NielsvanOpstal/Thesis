```python
def fix_ff(name):
    filled15 = pd.read_csv(name + "filled_data15.csv", index_col = 0)
    filled16 = pd.read_csv(name + "filled_data16.csv", index_col = 0)
    filled17 = pd.read_csv(name + "filled_data17.csv", index_col = 0)
    filled18 = pd.read_csv(name + "filled_data18.csv", index_col = 0)
    filled19 = pd.read_csv(name + "filled_data19.csv", index_col = 0)
    
    grid15 = pd.read_csv(name + "grid2015.csv")
    grid16 = pd.read_csv(name + "grid2016.csv")
    grid17 = pd.read_csv(name + "grid2017.csv")
    grid18 = pd.read_csv(name + "grid2018.csv")
    grid19 = pd.read_csv(name + "grid2019.csv")
    
    
    filled15 = filled15.join(grid15[["C28992R100", "geometry", "y", "BU_CODE"]])
    filled16 = filled16.join(grid15[["C28992R100", "geometry", "y", "BU_CODE"]])
    filled17 = filled17.join(grid15[["C28992R100", "geometry", "y", "BU_CODE"]])
    filled18 = filled18.join(grid15[["C28992R100", "geometry", "y", "BU_CODE"]])
    filled19 = filled19.join(grid15[["C28992R100", "geometry", "y", "BU_CODE"]])
    
    filled15.to_csv(name + "filled_data15.csv")
    filled16.to_csv(name + "filled_data16.csv")
    filled17.to_csv(name + "filled_data17.csv")
    filled18.to_csv(name + "filled_data18.csv")
    filled19.to_csv(name + "filled_data19.csv")
```

```python
df = pd.read_csv("pand_cleaned.csv", sep = ";")
df = gpd.GeoDataFrame(df)
df.geometry = df.geometry.apply(shapely.wkt.loads)
```

```python
df = df[df.bouwjaar < 2022]
```

```python
df.loc[df.bouwjaar < 1700, "bouwjaar"] = 1700
```

```python
df.loc[df.begindatumTijdvakGeldigheid.isna(), "begindatumTijdvakGeldigheid"] = df[df.begindatumTijdvakGeldigheid.isna()]["bouwjaar"].astype(int).astype(str)
```

```python
df["begindatumTijdvakGeldigheid"] =  pd.to_datetime(df['begindatumTijdvakGeldigheid'], errors='coerce')
```
